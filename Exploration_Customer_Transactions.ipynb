```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# if getting encoding error (error :- 'utf-8 code can't decode byte 0xa0 in position 0:') then use the below methode  
# df = pd.read_csv(filename, encoding = 'unicode_escape')
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
# drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df)
```

```python
# check for null values sum() methode can give you total no of null values present in each columns
pd.isnull(df).sum()
```

```python
df.shape
```

```python
# drop null values from dataframe
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']= df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shadi'})
```

```python
df.describe()
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

```python
df.columns
```

```python
ax = sns. countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)

# Add annotations to show the numbers on the graph
for index, row in sales_gen.iterrows():
    plt.text(row.name, row['Amount'] + 3, f"${row['Amount']:,.0f}", ha="center", va="bottom")
```

```python
ax = sns.countplot(x = 'Age Group', hue = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total amount vs age group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Create the bar plot with a rainbow color scheme
sns.barplot(x='Age Group', y='Amount', data=sales_age, hue='Age Group', palette='rainbow')

# Adjust the legend to avoid displaying it
plt.legend([],[], frameon=False)

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.show()
```

```python
# Total number of order from top 10 states

sales_state = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x= 'State', y='Orders',hue = 'State' ,palette='rainbow')
```

```python
# total amount/sales from top 10 states 
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(18,5)})
sns.set_palette("rainbow")
sns.barplot(data = sales_state, x = 'State', y = 'Amount',hue = 'State', palette='rainbow')
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc = {'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index= False)['Amount'].sum().sort_values(by = 'Amount')

sns.set(rc = {'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount', hue = 'Gender',palette='rainbow')
```

```python
sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(data = df, x = 'Occupation',hue = 'Occupation' ,palette='rainbow')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x= 'Occupation', y= 'Amount',hue = 'Occupation' ,palette='rainbow')
```

```python
sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(data = df, x = 'Product_Category',hue = 'Product_Category' ,palette='rainbow')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'],as_index=False).sum().sort_values(by = 'Amount', ascending= False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category', y = 'Amount', hue = 'Product_Category', palette='rainbow')
```

```python
sales_state = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID', y = 'Orders', hue = 'Product_ID' ,palette='rainbow')
```
